```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.python.keras.layers.core import Dense
from tensorflow.keras.models import Sequential
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from tensorflow.keras.utils import to_categorical
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

# User1 Flip 10%

```python
data = pd.read_csv('/content/drive/MyDrive/SwarmLearn/Cardio/Dataset/PC1.csv', header = None)
```

```python
X = data.iloc[:, :-1]
Y = data.iloc[:, -1]
```

```python
Y = Y.apply(lambda x: 1 if x != 0 else 0)
```

```python
y, y_flip = train_test_split(Y, test_size=0.1, random_state=42)
Y = pd.concat([y, y_flip], axis=0)
data = pd.concat([X, Y], axis = 1)
data.to_csv('/content/drive/MyDrive/SwarmLearn/Cardio/Flip_data/PC1.csv', index=False)
```
