# TH1 Lật data 10-20-30:

## User 1: 10%

```python
data = pd.read_csv('/content/drive/MyDrive/SwarmLearn/Fraunt/Dataset/SB19_CCFDUBL_BAL_P0_2C.csv')
```

```python
X = data.iloc[:, :-1]
Y = data.iloc[:, -1]
```

```python
y, y_flip = train_test_split(Y, test_size=0.1, random_state=42)
y_flip = y_flip.apply(lambda x: 1 if x == 0 else 0)
Y = pd.concat([y, y_flip], axis=0)
data = pd.concat([X, Y], axis = 1)
data.to_csv('/content/drive/MyDrive/SwarmLearn/Fraunt/Flip_data/SB19_CCFDUBL_BAL_P0_2C.csv', index=False)
```
